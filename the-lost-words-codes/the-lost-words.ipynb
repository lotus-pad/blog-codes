```python
import csv
```

```python
### saving data into csv as we go to avoid hitting limit

# here are all the things that is inside volumeInfo
#['title', 'subtitle', 'authors', 'publisher', 'publishedDate', 'description', 'industryIdentifiers', 'readingModes', 'pageCount', 'printType', 
# 'categories', 'averageRating', 'ratingsCount', 'maturityRating', 'allowAnonLogging', 'contentVersion', 'panelizationSummary', 'imageLinks', 
# 'language', 'previewLink', 'infoLink', 'canonicalVolumeLink']

output_file = "books.csv"

with open(output_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Author", "Published Date", "Categories", "Description", "ISBN_13", "Maturity Rating"])

books = []

for _, row in df.iterrows():
    title = row['Title']
    last_name = row['Last Name']

    query = f'intitle:"{title}" inauthor:"{last_name}"'

    params = {
        'q': query,
        'fields': 'items(volumeInfo)',
        'key': books_key
    }

    r = requests.get(url, params=params)

    if r.status_code == 429:
        print("Rate limit exceeded. Stopping search.")
        break

    if r.ok:
        data = r.json() # converting API response from JSON format into a python dictionary 
        items = data.get("items", []) 
        # This retrieves the "items" key from the data dictionary.
        # If "items" exists, it contains a list of book results from the search query.
        # If "items" does not exist (e.g., no books found), it defaults to an empty list [] to avoid errors.

        for item in items: ### what does items look like?
            info = item.get("volumeInfo", {})

            # Extract ISBN_13 if available
            isbn_13 = ""
            for identifier in info.get("industryIdentifiers", []):
                if identifier.get("type") == "ISBN_13":
                    isbn_13 = identifier.get("identifier", "")
                    break  # Stop after finding the first ISBN_13

            book_data = [
                info.get("title", ""),
                ", ".join(info.get("authors", [])),
                info.get("publishedDate", ""),
                ", ".join(info.get("categories", [])) if "categories" in info else "",
                info.get("description", ""),
                isbn_13,
                info.get("maturityRating", "")
            ]

            books.append(book_data)

            # Append to CSV file
            with open(output_file, mode="a", newline="", encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerow(book_data)

        print(f"Fetched data for {title} by {last_name}")

    else:
        print(f"Failed to fetch data for {title} by {last_name}")

    # Sleep for 1 second to avoid hitting the rate limit
    time.sleep(1)

print(f"Fetched data for {len(books)} books")
```
